### Description
load a .wav file of an MTD excerpt, compute an HCQT, and save it

```python
import os
import numpy as np
import librosa
import matplotlib.pyplot as plt
from hcqt import compute_hcqt, compute_efficient_hcqt
from tqdm.notebook import tqdm
```

```python
MTD_dir = "path_to_MTD"


wav_dir = os.path.join(MTD_dir, "data_AUDIO")

all_wavs = [f.split(".")[0] for f in os.listdir(wav_dir) if ".wav" in f]
all_wavs.sort()
```

```python
for file in tqdm(all_wavs):

    wav_in, fs_in = librosa.load(os.path.join(wav_dir, file+".wav"), mono=True)    
    feature_hcqt, fs_hcqt, hopsize_hcqt = compute_hcqt(wav_in,
                                                       fs=fs_in,
                                                       fmin = librosa.note_to_hz('C1'),
                                                       fs_hcqt_target=25,
                                                       bins_per_octave = 36,
                                                       num_octaves = 6,
                                                       num_harmonics = 5,
                                                       num_subharmonics = 1,
                                                       center_bins = True)    
    np.save(file=os.path.join("data", "hcqt", file + ".npy"), arr=feature_hcqt)
```
